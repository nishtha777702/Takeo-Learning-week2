```python
# Task
# load/create the dataset into a DataFrame and display the first row

import pandas as pd
import numpy as np

#load the dataset 
# Load the dataset
data = {
 'ID': [1, 2, 3, 4, 5, 6, 7, 7],
 'Name': ['Alice', 'Bob', 'Charlie', np.nan, 'Eve', 'Frank',
'Grace', 'Grace'],
 'Age': [25, np.nan, 30, 45, 'unknown', 35, 30, 30],
 'Salary': [50000, 60000, np.nan, 70000, 80000, 'not available',
70000, 70000],
 'Department': ['HR', 'Finance', 'HR', 'IT', np.nan, 'Finance',
'IT', 'IT']
}
df = pd.DataFrame(data)
df
# Display the first few rows
#df.head(7)
df.describe() # fucntions that gives you a quick statistical summary of your Dataframe -mainly for numeric columns
#df['head'].fillna('Unknown' ,inplace='TRUE')
```

```python
# identify the missing values
#To find and display the count of missing value in each columns

#check for missing value 

missing_values = df.isnull().sum()
missing_values
```

```python
# HAndle missing value
# ðŸ”¹ Step 1: Fill missing 'Name' values with 'Unknown'
df['Name'].fillna('Unknown', inplace=True)

# Step 2: Convert 'Age' and 'Salary' to numeric (invalid entries become NaN)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

#  ðŸ”¹ Step 3: Fill missing 'Department' values with 'Unknown'
df['Department'].fillna('Unknown', inplace=True)

# # Display the first 8 rows to confirm changes
print(df.head(8))
```

```python
# Remove Duplicate rows based on ID

df = df.drop_duplicates(subset= 'ID')  # the subset parameters allows you to specify which columns you want to check fro duplicates like in this case "ID'
df.head(10)
```

```python
from scipy import stats

# Convert 'Salary' column to numeric, coercing errors to NaN
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Calculate Z-scores after handling invalid salary entries
df['Salary_zscore'] = stats.zscore(df['Salary'].fillna(df['Salary'].mean()))

# Define a threshold for outliers
threshold = 3

# Identify outliers
df_outliers = df[df['Salary_zscore'].abs() > threshold] # abs() function calculate absoulte value of each z- score in the salary_zscore column

# Remove outliers
df_cleaned = df[df['Salary_zscore'].abs() <= threshold]

# Display the DataFrames for outliers and cleaned data
df_outliers, df_cleaned
```

```python
# validate data
valid_data = df_cleaned[df_cleaned['Age'] > 0]
valid_data = valid_data[valid_data['Salary'] > 0]
# Display the validated data
valid_data
```
